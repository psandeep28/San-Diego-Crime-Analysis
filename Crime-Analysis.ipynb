```python
income = pd.read_csv('income_SD.csv')
```

```python
# Remove commas then convert to integer format
income['Population'] = income['Population'].astype(str).str.replace(',', '').astype(int)
income['Number of Households'] = income['Number of Households'].astype(str).str.replace(',', '').astype(int)

#Remove dollar sign and commas then convert to float format
income['Median Income'] = income['Median Income'].astype(str).str.replace('$', '').str.replace(',', '').astype(float)
income['Average Income'] = income['Average Income'].astype(str).str.replace('$', '').str.replace(',', '').astype(float)
```

```python
healthcare_df = pd.read_csv('Healthcare.csv')
library_locations_df = pd.read_csv('Library_locations.csv')
crime_data_df = pd.read_csv('ARJIS_Public_Crime_Data_w__Day_of_Week_20240222-2.csv')
income_df = pd.read_csv('Income_SD.csv')
```

```python
def categorize_income(income):
    if income <= 50000:
        return 'Low'
    elif income <= 100000:
        return 'Medium'
    elif income <= 150000:
        return 'High'
    else:
        return 'Very High'
healthcare_df.dropna(subset=['DBA_ZIP_CODE'], inplace=True)
healthcare_df['DBA_ZIP_CODE'] = healthcare_df['DBA_ZIP_CODE'].astype(int).astype(str).str.pad(width=5, fillchar='0')

library_locations_df.dropna(subset=['zip'], inplace=True)
library_locations_df['zip'] = library_locations_df['zip'].astype(int).astype(str).str.pad(width=5, fillchar='0')

crime_data_df.dropna(subset=['ZIP Code'], inplace=True)
crime_data_df['ZIP Code'] = crime_data_df['ZIP Code'].astype(int).astype(str).str.pad(width=5, fillchar='0')

income_df.dropna(subset=['ZipCode'], inplace=True)
income_df['ZipCode'] = income_df['ZipCode'].astype(int).astype(str).str.pad(width=5, fillchar='0')

healthcare_df['DBA_ZIP_CODE'] = healthcare_df['DBA_ZIP_CODE'].astype(str).str.split('.').str[0].str.pad(width=5, fillchar='0')
library_locations_df['zip'] = library_locations_df['zip'].astype(str).str.pad(width=5, fillchar='0')
crime_data_df['ZIP Code'] = crime_data_df['ZIP Code'].astype(str).str.split('.').str[0].str.pad(width=5, fillchar='0')
income_df['ZipCode'] = income_df['ZipCode'].astype(str).str.pad(width=5, fillchar='0')


# Clean and prepare healthcare data
healthcare_counts = healthcare_df.groupby('DBA_ZIP_CODE').size().reset_index(name='Healthcare_Count')

# Clean and prepare library data
library_counts = library_locations_df.groupby('zip').size().reset_index(name='Library_Count')

# Prepare crime data
crime_data_df['ZIP Code'] = crime_data_df['ZIP Code'].fillna(0).astype(int).astype(str).str.pad(width=5, side='left', fillchar='0')
crime_counts = crime_data_df.groupby('ZIP Code').size().reset_index(name='Crime_Count')

# Clean and prepare income data
income_df['Median Income'] = income_df['Median Income'].replace('[\$,]', '', regex=True).astype(float)
income_df['ZipCode'] = income_df['ZipCode'].astype(str).str.pad(width=5, side='left', fillchar='0')

# Merge datasets
merged_df_1 = pd.merge(crime_counts, healthcare_counts, left_on='ZIP Code', right_on='DBA_ZIP_CODE', how='left')
merged_df_2 = pd.merge(merged_df_1, library_counts, left_on='ZIP Code', right_on='zip', how='left')
final_df = pd.merge(merged_df_2, income_df[['ZipCode', 'Median Income']], left_on='ZIP Code', right_on='ZipCode', how='inner')

# Clean up the merged DataFrame
final_df.drop(columns=['DBA_ZIP_CODE', 'zip', 'ZipCode'], inplace=True)
final_df.fillna({'Healthcare_Count': 0, 'Library_Count': 0}, inplace=True)

median_income_median = final_df['Median Income'].median()
final_df['Income Category'] = final_df['Median Income'].apply(lambda x: 'High' if x > median_income_median else 'Low')

# Categorize presence of healthcare and libraries
final_df['Has Healthcare'] = final_df['Healthcare_Count'].apply(lambda x: 'Yes' if x > 0 else 'No')
final_df['Has Library'] = final_df['Library_Count'].apply(lambda x: 'Yes' if x > 0 else 'No')

final_df['Facilities Present'] = final_df.apply(lambda x: 'Healthcare & Library' if x['Has Healthcare'] == 'Yes' and x['Has Library'] == 'Yes' 
                                                else ('Healthcare Only' if x['Has Healthcare'] == 'Yes' 
                                                      else ('Library Only' if x['Has Library'] == 'Yes' else 'None')), axis=1)

final_df['Income Level'] = final_df['Median Income'].apply(categorize_income)
```

```python
df1_cleaned['ZIP Code'] = df1_cleaned['ZIP Code'].astype(int) 
merged_data = pd.merge(df1_cleaned, income, left_on='ZIP Code', right_on='ZipCode', how='inner')
merged_data
```

```python
crime_count_by_zip = merged_data.groupby('ZIP Code')['Crime Category'].count().reset_index(name='Total Crimes')
crime_count_by_zip.sort_values(by = 'Total Crimes', ascending= False)
```

```python
income_crime_data = pd.merge(crime_count_by_zip, income, left_on='ZIP Code', right_on='ZipCode')
income_crime_data
```

```python
# Categorizing income levels
income_brackets = pd.cut(income_crime_data['Median Income'], bins=[0, 50000, 100000, 150000, np.inf], labels=['Low', 'Medium', 'High', 'Very High'])
income_crime_data['Income Bracket'] = income_brackets
income_crime_data
```

In order to better group the income levels we are working with, we decided to break each aspect into 4 groups: "Low, Medium, High, Very High". This will give us a better understanding of the income from a categorical sense. 

```python
merged_data['Income Bracket'] = income_brackets
crime_category_income = merged_data.groupby(['Income Bracket', 'Crime Category']).size().reset_index(name='Counts')

sns.catplot(x='Income Bracket', y='Counts', hue='Crime Category', data=crime_category_income, kind='bar')
plt.title('Crime Category Distribution Across Income Brackets')
plt.xlabel('Income Bracket')
plt.ylabel('Crime Counts')
plt.show()
```

The chart suggests a trend where burglary and assault are more frequent in lower income brackets, indicating these crimes might be more prevalent in poorer areas. As income levels increase, the frequency of these crimes appears to decrease. However, DUIs and drug/alcohol violations seem consistent across all income levels, implying these crimes are not as influenced by economic status.

```python
income_crime_data.columns
income_crime_data = income_crime_data[income_crime_data['Total Crimes'] < 4500] # outlier
income_crime_data = income_crime_data[income_crime_data['Median Income'] < 212500] # outlier

changed_income_crime_data = income_crime_data.rename(columns={'Total Crimes': 'TotalCrimes', 'Median Income': 'MedianIncome'})
income_crime_outcome, income_crime_predictors = patsy.dmatrices('TotalCrimes ~ MedianIncome', changed_income_crime_data)
income_crime_model = sm.OLS(income_crime_outcome, income_crime_predictors)
income_crime_results = income_crime_model.fit()

print(income_crime_results.summary())
```

R-squared value: The R-squared value of 0.182 indicates that the model can explain around 12.2% of the variation in Total Crimes. However, the Adjusted R-squared value of 0.104 is slightly lower, suggesting that there might be some overfitting or that additional relevant variables are not included in the model.

Median Income: The coefficient for Median Income is -0.0063, with a p-value of 0.000, which is highly statistically significant. This negative coefficient suggests that as Median Income increases, Total Crimes tends to decrease.

Omnibus: The Omnibus test statistic (2.234) has a p-value of 0.327, which is not statistically significant at the 5% level.

Jarque-Bera (JB): The Jarque-Bera test statistic (1.606) has a p-value of 0.448, which is not statistically significant at the 5% level. This suggests that the residuals are normally distributed, satisfying one of the assumptions of the OLS regression model.

Condition Number: The large condition number (3.43e+05) suggests the presence of multicollinearity in the model.

Overall, while the model finds a statistically significant negative relationship between Median Income and Total Crimes. However, the low R-squared values, non-normality of residuals, and potential multicollinearity issues indicate that the model may not capture all the relevant factors influencing Total Crimes.

H1: The mean crime rate in low-income areas is not equal to the mean crime rate in high-income areas.

```python
low_income_df = income_crime_data[income_crime_data['Income Bracket'] == "Low"]
high_income_df = income_crime_data[income_crime_data['Income Bracket'] == "High"]
```

```python
from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(low_income_df['Total Crimes'], high_income_df['Total Crimes'])
t_stat, p_value
```

T-statistic (0.037): A t-statistic near 0 suggests that there's little difference between the group means. In this case, the very low t-statistic (0.037) indicates that the mean difference in total crimes between low-income and high-income areas is small relative to the variability in crime counts within those areas.

P-value (0.970): A p-value greater than 0.05 means there's insufficient evidence to reject the null hypothesis. Therefore, we do not have sufficient evidence to say there is a significant difference in the mean total crimes between low-income and high-income areas based on this analysis.

## ANOVA Analysis

To perform the ANOVA Analysis we followed these steps: 
1. Calculate the crime rate per 1,000 people for each ZIP code across all income brackets.
2. Group these rates by their respective income brackets.
3. Use the ANOVA test to compare the mean crime rates across these groups.
4. Interpret the F-statistic and p-value to understand if there are statistically significant differences in crime rates among the different income brackets.

```python
from scipy.stats import f_oneway

# Calculating crime rate per 1,000 people for the entire dataset
income_crime_data['Crime Rate per 1000'] = income_crime_data['Total Crimes'] / income_crime_data['Population'] * 1000

# Grouping the data by income bracket
grouped_data = income_crime_data.groupby('Income Bracket')['Crime Rate per 1000']

# Preparing lists of crime rates for each income bracket
low_crime_rates = income_crime_data[income_crime_data['Income Bracket'] == "Low"]['Crime Rate per 1000']
medium_crime_rates = income_crime_data[income_crime_data['Income Bracket'] == "Medium"]['Crime Rate per 1000']
high_crime_rates = income_crime_data[income_crime_data['Income Bracket'] == "High"]['Crime Rate per 1000']
very_high_crime_rates = income_crime_data[income_crime_data['Income Bracket'] == "Very High"]['Crime Rate per 1000']

# Performing ANOVA test
f_stat, p_value = f_oneway(low_crime_rates, medium_crime_rates, high_crime_rates, very_high_crime_rates)

# Calculating mean crime rates for each income bracket
mean_rates = grouped_data.mean()

f_stat, p_value, mean_rates
```

F-statistic (4.367): A higher F-statistic suggests a greater variance between the means of the groups than within the groups, indicating that at least one group's mean significantly differs from the others.

P-value (0.0065): A p-value of approximately 0.0194 (less than the conventional threshold of 0.05) suggests that we can reject the null hypothesis with a confidence level of about 98%, indicating that there is a statistically significant difference in crime rates among the different income brackets.

According to mean crime rates, the income categories with the highest rate of crime per 1,000 persons are those in the medium range; the income brackets with the lowest rate of crime are those in the high, low, and very high ranges. Given that the Medium bracket has a greater crime rate than the Low bracket, this trend may appear illogical.
